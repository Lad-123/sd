```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data pre-processing or Preparation of the data

```python
X = iris['data']
X [0:10]
```

```python
y = iris['target']
y[0:10]
```

### Visualizing

```python
plt.plot(X[: ,0]); # Corrsponding to sepal length
```

```python
plt.plot(X[: ,1]); # Corrsponding to sepal width
```

```python
plt.plot(X[: ,2]); # Corrsponding to petal length
```

```python
plt.plot(X[: ,3]); # Corrsponding to petal width
```

```python
plt.plot(X[: ,0], label = 'Sepal Length');
plt.plot(X[: ,1], label = 'Sepal Width');
plt.plot(X[: ,2], label = 'Petal Length');
plt.plot(X[: ,3], label = 'Petal Width');
plt.legend();
```

### Splitting the Data : To train and test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 100)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train
```

### Building the model

```python
from sklearn.tree import DecisionTreeClassifier

dt_1 = DecisionTreeClassifier() # DecisionTreeClassifier is a class and needs to be called using an object dt_1
```

```python
# Training the model

dt_1 = dt_1.fit(X_train, y_train)
```

```python
dt_1
```

### Prediction

```python
y_pred = dt_1.predict(X_test)
y_pred
```

```python
y_test
```

### Finding the performance measures

```python
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, roc_auc_score

cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot = True);
```

```python
report = classification_report(y_test, y_pred)
print ('The classification report:\n', report)
```

```python
fpr, tpr, _ = roc_curve(y_test, y_pred)
plt.plot(fpr, tpr)
```

ROC Curve is applicable for binary classification. Since this program has 3 classifications, hence, roc_curve doesn't work

### Visualization of the tree

```python
from sklearn.tree import plot_tree
plt.figure(figsize = (15, 15))
plot_tree (dt_1, filled = True);
```

gini is like entropy which measures the randomness of the model. Lesser the gini value, better is the model. 

 $gini = 1 - \sum (p^2)$, p = probability of no of observation under a particular class

```python
gini = 1 - (39 / 120) ** 2 - (44 / 120) ** 2 - (37 / 120) ** 2
gini
```

```python
## Displaying the rules to be followed

from sklearn.tree import export_text

text = export_text(dt_1, feature_names = iris['feature_names'])
print (text)
```

### Modifying the model

```python
dt_2 = DecisionTreeClassifier(criterion = 'entropy', max_depth = 4)

dt_2 = dt_2.fit(X_train, y_train)
y_pred_2 = dt_2.predict(X_test)
cm_2 = confusion_matrix(y_test, y_pred_2)
sns.heatmap(cm_2, annot = True);
```

```python
plt.figure(figsize = (15, 15))
plot_tree(dt_2, filled = True);
```
