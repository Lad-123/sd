### Gradient Descent, Errors

### Accessing the Dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Accessing the Olympics dataset

ol = pd.read_csv('olympics.csv')
```

```python
ol
```

```python
# Checking for the null values

ol.isna().sum()
```

### Data Pre-processing

```python
# Defining the features

X = ol['Year']
X
```

```python
# Defining the target

y = ol.drop(['Year'], axis = 1)
y
```

### Plotting

```python
plt.figure (figsize = (10, 10))
plt.scatter(X, y);
```

### Designing a Simple LR model

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression() # lr is the object for the class Linear Regression
lr.fit(X.to_numpy().reshape (-1, 1), y)
```

```python
plt.figure (figsize = (10, 10))
plt.scatter(X, y);
plt.plot(X, lr.predict(X.to_numpy().reshape(-1, 1)));
```

**Gradient Descent** 

    The first order partial derivative of the function gives us a gradient descent.

**We create a function by the name Loss function (L), which is nothing but the sum of squares of the Error (SSE). We then use the partial derivative to find the local minima to determine the value at which the slope is 0**

**There are 2 types of Errors associated with the Gradient Descent model which is mentioned below:**

1. Bias Error (Error related to the model). The result are generally underfitting.
2. Variance Error (Error related to the dataset). The results are generally overfitting.

**The gradient descent model underfits the data points. We therefore try to convert the linear equation into a polynomial equation of degree 2**

### Polynomial Regression

```python
# Creating a polynomial model of degree 2

from sklearn.preprocessing import PolynomialFeatures

p = PolynomialFeatures(degree = 2)
X_2 = p.fit_transform(X.to_numpy().reshape (-1, 1)) # This transforms the features value
X_2
```

```python
lr2 = LinearRegression()
p_2 = lr2.fit(X_2, y) # We are trying to use the Linear Regression to the equation (1 + X + X^2)
```

```python
# Plotting the equation showing polynomial equation of degree 2

plt.figure (figsize = (10, 10))
plt.scatter (X, y)
plt.plot (X, lr.predict(X.to_numpy().reshape (-1, 1)), label = 'Linear Regression')
plt.plot(X, p_2.predict (X_2), label = 'Polynomial Regression of Degree 2');
plt.legend();
```

**By changing the bias from a linear to a polynomial model, the error related to bias has reduced. We try to reduce the error further by changing to polynomial function to degree 3**

```python
# Creating a polynomial model of degree 3

from sklearn.preprocessing import PolynomialFeatures

p_3 = PolynomialFeatures(degree = 3)
X_3 = p_3.fit_transform(X.to_numpy().reshape (-1, 1)) # This transforms the features value
X_3
```

```python
lr3 = LinearRegression()
p_3 = lr3.fit(X_3, y) # We are trying to use the Linear Regression to the equation (1 + X + X^2 + X^3)
```

```python
# Plotting the equation showing polynomial equation of degree 3

plt.figure (figsize = (10, 10))
plt.scatter (X, y)
plt.plot (X, lr.predict(X.to_numpy().reshape (-1, 1)), label = 'Linear Regression')
plt.plot(X, p_2.predict (X_2), label = 'Polynomial Regression of Degree 2');
plt.plot(X, p_3.predict (X_3), label = 'Polynomial Regression of Degree 3');
plt.legend();
```

**When we go on increasing the complexity of the equation, i.e. increasing the degree of the equation, we try to overfit the data.**

**Whenever we try to reduce the Bias error, the Variance error increases and vice-versa. We therefore try to come to an optimum value for both the errors and it is called "Bias - Variance Trade Off".**
