### Data Preprocessing

```python
# Target

y = ipl['SOLD PRICE']
y
```

```python
# Features

X = ipl.drop(['SOLD PRICE'], axis = 1)
X
```

We will be dropping the Player's name as it is a categorical value. Along with that we will be dropping the Sl No, Team.

```python
# Drop irrelevant features

X_1 = X.drop(['Sl.NO.','PLAYER NAME', 'TEAM'],axis = 1)
X_1.head()
```

```python
X_1['AGE'].unique()
```

```python
X['COUNTRY'].unique()
```

```python
X['PLAYING ROLE'].unique()
```

```python
X['CAPTAINCY EXP'].unique()
```

```python
# Convert categorical to numerical using one hot encoding

X_1 = pd.get_dummies(X_1,columns = ['AGE', 'COUNTRY', 'PLAYING ROLE', 'CAPTAINCY EXP'])
```

```python
X_1
```

```python
X_1.iloc[0:10,0:15]
```

```python
X_1.iloc[0:10,15:]
```

```python
import statsmodels.api as sm
X_1 = sm.add_constant(X_1)
X_1
```

```python
X_1.shape
```

### Splitting data to train and test

```python
from sklearn.model_selection import train_test_split

X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(
X_1, y, test_size = 0.2, random_state = 10)
```

```python
X_train_1.shape, X_test_1.shape, y_train_1.shape, y_test_1.shape
```

```python
X_train_1
```

```python
X_test_1
```

### Building the model

```python
mlr_1 = sm.OLS(y_train_1, X_train_1)
```

```python
mlr_1 = mlr_1.fit()
```

```python
mlr_1.params
```

### Diagnosing the model

```python
mlr_1.summary2()
```

Note - Only ODI_WKTS and BASE PRICE are relevant features.

Assumption - All the features for Multiple Linear Regression has to be independent. Since, the features are not independent, the statistics achieved is affected.

### Multicollinearity

VIF = Variance Inflation Factor (1 - (1/R^2))

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def var_inf_factor(data):
    vif = pd.DataFrame()
    vif['Feature'] = data.columns
    vif['VIF_Value'] = [variance_inflation_factor(data.values, i) for i in range(data.shape[1])]
    print (vif)
```

```python
var_inf_factor(X_1)
```
